```python
import pandas as pd
import numpy as np

df = pd.read_csv("file:///C:/Users/kwasi/Documents/Python%20for%20Analytics/EDdata_new.csv")

df["wait_time"] = 0
n_pat = len(df)
df_reg = df["reg_sec"]
df_tri = df["triage_sec"]
df_wait = df["wait_time"]
df_wait = (24*60*60) + df_tri - df_reg
filt_1 = df["triage_sec"] > df["reg_sec"]
df_wait[filt_1] = df_tri[filt_1] - df_reg[filt_1]
df["wait_time"] = df_wait
df["wait_min"] = df_wait/60
```
